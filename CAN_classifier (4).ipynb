```python
import zipfile
import os
import pandas as pd
from pathlib import Path

# Extrage ZIP-ul
with zipfile.ZipFile('datasets_with_attacks.zip', 'r') as zip_ref:
    zip_ref.extractall('datasets_with_attacks')

print("Conținutul folderului după extragere:")
print(os.listdir('datasets_with_attacks'))

# Funcție pentru găsirea automată a fișierelor CSV relevante
def find_vehicle_csv_files(base_directory):
    """Găsește automat fișierele CSV pentru vehicule"""
    csv_files = {}

    # Caută în toate subdirectoarele
    for root, dirs, files in os.walk(base_directory):
        for file in files:
            if file.endswith('.csv') and 'Vehicle' in file:
                # Încearcă să determine numele vehiculului din numele fișierului
                if 'Vehicle A' in file:
                    vehicle_name = 'Vehicle A'
                elif 'Vehicle B' in file:
                    vehicle_name = 'Vehicle B'
                elif 'Vehicle C' in file:
                    vehicle_name = 'Vehicle C'
                else:
                    # Dacă nu găsim un pattern standard, folosim numele fișierului
                    vehicle_name = file.split(' - ')[0] if ' - ' in file else file.replace('.csv', '')

                csv_files[vehicle_name] = os.path.join(root, file)
                print(f"Găsit: {vehicle_name} -> {os.path.join(root, file)}")

    return csv_files

# Găsește fișierele
vehicle_files = find_vehicle_csv_files('datasets_with_attacks')
extra_file_path = os.path.join('datasets_with_attacks', 'Datasets with attacks', 'Fuzzing', 'Vehicle A',
                               'Vehicle A - Replay 0x181, Fuzzing 0x244 - IDs_Datafield_Classification.csv')

# Verifică dacă fișierul există și nu este deja inclus
if os.path.isfile(extra_file_path) and 'Vehicle A - Replay 0x181, Fuzzing 0x244' not in vehicle_files:
    vehicle_files['Vehicle A - Replay 0x181, Fuzzing 0x244'] = extra_file_path
    print(f"\nFișier suplimentar adăugat manual: {extra_file_path}")

if not vehicle_files:
    print("Nu s-au găsit fișiere CSV cu 'Vehicle' în nume. Să căutăm toate fișierele CSV:")

    # Caută toate fișierele CSV
    all_csv_files = []
    for root, dirs, files in os.walk('datasets_with_attacks'):
        for file in files:
            if file.endswith('.csv'):
                full_path = os.path.join(root, file)
                all_csv_files.append(full_path)
                print(f"CSV găsit: {full_path}")

    print(f"\nTotal {len(all_csv_files)} fișiere CSV găsite.")

    if all_csv_files:
        print("\nÎncerc să încarc primul fișier pentru a vedea structura:")
        sample_file = all_csv_files[0]
        print(f"Încarcă: {sample_file}")

        # Încearcă diferite delimitatoare
        for delimiter in [';', ',', '\t']:
            try:
                df_sample = pd.read_csv(sample_file, delimiter=delimiter, nrows=5)
                print(f"\nCu delimitatorul '{delimiter}':")
                print(f"Forme: {df_sample.shape}")
                print(f"Coloane: {list(df_sample.columns)}")
                print("Primele 3 rânduri:")
                print(df_sample.head(3))
                break
            except Exception as e:
                print(f"Eșuat cu delimitatorul '{delimiter}': {e}")

# Dacă am găsit fișiere vehicle
if vehicle_files:
    print(f"\nÎncarc {len(vehicle_files)} fișiere vehicle...")
    dfs = {}

    for vehicle_name, file_path in vehicle_files.items():
        try:
            print(f"\nÎncarc {vehicle_name} din {file_path}")

            # Încearcă diferite delimitatoare
            df = None
            for delimiter in [';', ',', '\t']:
                try:
                    df = pd.read_csv(file_path, delimiter=delimiter)
                    print(f"Succes cu delimitatorul '{delimiter}'")
                    break
                except:
                    continue

            if df is not None:
                # Redenumește coloanele
                df.columns = [f'Column{i}' for i in range(len(df.columns))]
                dfs[vehicle_name] = df
                print(f"Forme dataset: {df.shape}")
                print(f"Coloane: {list(df.columns)}")
            else:
                print(f"Nu s-a putut încărca {vehicle_name}")

        except Exception as e:
            print(f"Eroare la încărcarea {vehicle_name}: {e}")

    print(f"\nÎncărcate cu succes: {len(dfs)} dataset-uri")
    for vehicle, df in dfs.items():
        print(f"  {vehicle}: {df.shape}")
```

```python
# Function to print proportion of 0s and 1s in Column13
def print_attack_proportions(df, vehicle_name):
    if df is not None:
        counts = df['Column13'].value_counts(normalize=True)
        print(f"\n{vehicle_name} - Proportion of labels in Column13:")
        print(counts)
    else:
        print(f"\n{vehicle_name} dataframe not loaded.")
```

```python
import numpy as np

def downsample_no_attack(df, label_col='Column13', no_attack_label=0, fraction=0.7):
    if df is None:
        return None
    no_attack_rows = df[df[label_col] == no_attack_label]
    attack_rows = df[df[label_col] != no_attack_label]

    # Randomly sample 60% (i.e., keep 60%) of no-attack rows
    no_attack_downsampled = no_attack_rows.sample(frac=1 - fraction, random_state=42)

    # Combine with all attack rows
    downsampled_df = pd.concat([no_attack_downsampled, attack_rows]).reset_index(drop=True)
    return downsampled_df

df_a = downsample_no_attack(df_a)
df_b = downsample_no_attack(df_b)
df_c = downsample_no_attack(df_c)

# Verify new proportions and counts
def print_info(df, name):
    if df is not None:
        total = len(df)
        counts = df['Column13'].value_counts(normalize=True)
        print(f"\n{name}:")
        print(f"Total entries: {total}")
        print(f"Label proportions:\n{counts}")
    else:
        print(f"{name} dataframe not loaded.")

print_info(df_a, "Vehicle A")
print_info(df_b, "Vehicle B")
print_info(df_c, "Vehicle C")
```

```python
df_all_3 = pd.concat([df_a, df_b, df_c], ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# Shuffle the data
df_all_3 = df_all_3.sample(frac=1, random_state=42).reset_index(drop=True)

# Features and target
X = df_all_3.drop(columns=['Column13'])
y = df_all_3['Column13']

# Split data 80-20 stratified to keep label proportions
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

lr = LogisticRegression(random_state=42, max_iter=2000)
```

```python
lr.fit(X_train_scaled, y_train)

# Predict on test data
y_test_pred = lr.predict(X_test_scaled)

# Print evaluation metrics
print(classification_report(y_test, y_test_pred))

precision = precision_score(y_test, y_test_pred, average='binary')
recall = recall_score(y_test, y_test_pred, average='binary')
f1 = f1_score(y_test, y_test_pred, average='binary')
accuracy = accuracy_score(y_test, y_test_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")
print(f"Accuracy:  {accuracy:.4f}")
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf.fit(X_train_scaled, y_train)

# Predict and evaluate
y_test_pred = rf.predict(X_test_scaled)

print(classification_report(y_test, y_test_pred))
```
